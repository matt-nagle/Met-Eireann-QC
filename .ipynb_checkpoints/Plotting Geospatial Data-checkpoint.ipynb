```python
import geopandas
import pandas
import numpy as np
```

```python
# Import and define the boundaries for the counties of the island of Ireland
path_to_ROI_boundaries_data = r"/home/matt/Documents/Met Eireann (Linux)/01. Boundaries/Republic of Ireland Counties - shp/7829fd91-e0c3-4246-8fb5-100f3b62f8272020329-1-8mtzfa.iikfy.shp"
path_to_NI_boundaries_data = r"/home/matt/Documents/Met Eireann (Linux)/01. Boundaries/Northern Ireland Counties - shp/OSNI_Open_Data_-_50K_Boundaries_-_NI_Counties.shp"

ROI_counties = geopandas.read_file(path_to_ROI_boundaries_data)
NI_counties = geopandas.read_file(path_to_NI_boundaries_data)
```

```python
# Note that we will use the Irish Grid projection so that we can get accurate distances
ROI_counties = ROI_counties.to_crs({'init': 'epsg:29902'}) # Convert maps to the same coordinate reference system
NI_counties = NI_counties.to_crs({'init': 'epsg:29902'})  # Convert maps to the same coordinate reference system

# Plot counties
ax = ROI_counties["geometry"].plot()
NI_counties["geometry"].plot(ax=ax)
```

```python
# Import Raw Wow data for the month of May
path_to_raw_wow_data = r"/home/matt/Documents/Met Eireann (Linux)/WOW_May_2021.csv"

raw_wow = pandas.read_csv(path_to_raw_wow_data, dtype = {"Ground State":"string"})
```

```python
# Define a dataframe for the Rain Accumulation data
rain_accumulation = raw_wow[['Id', 'Site Id', 'Longitude', 'Latitude', 'Report Date / Time','Rainfall Accumulation']]

# Remove any missing values
rain_accumulation_complete = rain_accumulation.dropna()
```

```python
# Define a dataframe for the Air Temperature data
temperature = raw_wow[['Id', 'Site Id', 'Longitude', 'Latitude', 'Report Date / Time','Air Temperature']]

# Remove any missing values
temperature_complete = temperature.dropna()
```

```python
def plot_wow_data(date_of_interest, time_of_interest, type_of_plot = {}):
    """
    This function takes the date and time of interest as well as the type of plot desired and plots the results.
    
    Note: 
    date_of_interest must be of the format "YYYY-MM-DD"
    time_of_interest must be of the 24 hour format "HH" and will plot data from the times between HH and HH+1
    type_of_plot must be either "Rainfall Accumulation" or "Air Temperature"
    """
    
    # define date and hour of interest
    date_and_time = date_of_interest + " " + time_of_interest
    

    
    if type_of_plot == "Rainfall Accumulation":
        # Isolate the data at that date and time
        data_of_interest = [date_and_time in date for date in rain_accumulation_complete["Report Date / Time"]]
        rain_accumulation_complete_time_of_int = rain_accumulation_complete.loc[data_of_interest,]
        
        # In the rainfall accumulation case we want to keep only the last observation before the hour of interest is finished
        rain_accumulation_complete_time_of_int_last_obs = rain_accumulation_complete_time_of_int.drop_duplicates(
            subset="Site Id", keep="last")

        # Convert data of interest into a GeoDataFrame for plotting
        gdf = geopandas.GeoDataFrame(rain_accumulation_complete_time_of_int_last_obs, 
                                     geometry=geopandas.points_from_xy(
                                         rain_accumulation_complete_time_of_int_last_obs.Longitude, 
                                         rain_accumulation_complete_time_of_int_last_obs.Latitude))
        
        gdf.crs = {"init":"epsg:4326"} # initialise the dataframe to have a crs
        gdf = gdf.to_crs({'init': 'epsg:29902'}) # convert crs to Irish Grid Projection
        
    elif type_of_plot == "Air Temperature":
        # Isolate the data at that date and time
        data_of_interest = [date_and_time in date for date in temperature_complete["Report Date / Time"]]
        temperature_complete_time_of_int = temperature_complete.loc[data_of_interest,]
        
        # In the temperature case we want to keep only the last observation before the hour of interest is finished
        temperature_complete_time_of_int_last_obs = temperature_complete_time_of_int.drop_duplicates(
            subset="Site Id", keep="last")

        # Convert data of interest into a GeoDataFrame for plotting
        gdf = geopandas.GeoDataFrame(temperature_complete_time_of_int_last_obs, 
                                     geometry=geopandas.points_from_xy(temperature_complete_time_of_int_last_obs.Longitude, 
                                                                       temperature_complete_time_of_int_last_obs.Latitude))
        
        gdf.crs = {"init":"epsg:4326"} # initialise the dataframe to have a crs
        gdf = gdf.to_crs({'init': 'epsg:29902'}) # convert crs to Irish Grid Projection
    
    
    # Plot county borders for the Island of Ireland
    ax = ROI_counties["geometry"].plot(figsize=(10,15), edgecolors="grey", color="w")
    NI_counties["geometry"].plot(ax=ax, edgecolors="grey", color="w")

    ax.set_title(date_and_time + "am - " + type_of_plot)    
    
    # Plot transparent markers
    gdf.plot(type_of_plot, ax=ax, markersize=400, legend = True, alpha = 0.5)

    # Annotate each marker with the Rainfall Accumulation to the nearest whole number
    for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf[type_of_plot]):
        ax.annotate(round(label), xy=(x, y), verticalalignment='center', horizontalalignment='center', 
                    weight = "bold", fontsize = "large")
        
    # Annotate each marker with the Rainfall Accumulation to the nearest whole number
    #make circle from point
    new_gdf = gdf.copy()
    new_gdf['geometry'] = new_gdf['geometry'].buffer(40000)

    new_gdf["geometry"].plot(ax=ax, alpha = 0.2)
```

```python
plot_wow_data(date_of_interest="2021-05-10", time_of_interest="12", type_of_plot="Rainfall Accumulation")
```

```python
plot_wow_data(date_of_interest="2021-05-05", time_of_interest="12", type_of_plot="Air Temperature")
```
